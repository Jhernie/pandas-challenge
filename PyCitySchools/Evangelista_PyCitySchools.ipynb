```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data)
student_data_df = pd.read_csv(student_data)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])

#####______summary attributes for data frame_____###
#school_data_complete_df.head()
#school_data_complete_df.shape
school_data_complete_df.columns
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = school_data_complete_df["school_name"].unique()
total_num_students = len(school_data_complete_df["student_name"])
total_budget = sum(school_data_complete_df["budget"].unique())
avg_math_score = school_data_complete_df["math_score"].mean()
avg_read_score = school_data_complete_df["reading_score"].mean()
passing_math = len(school_data_complete_df[school_data_complete_df["math_score"] >= 70])
passing_read = len(school_data_complete_df[school_data_complete_df["reading_score"] >= 70])
percent_passing_math = ((passing_math)/total_num_students)*100
percent_passing_read = ((passing_read)/total_num_students)*100
overall_passing = len(school_data_complete_df[(school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)])
percent_overall_passing = (overall_passing/total_num_students)*100

district_summary_df = pd.DataFrame(
    { 
        "Total Number of Schools": len(total_schools),
        "Total Number of Students": [total_num_students],
        "Total Budget": f'${total_budget}',
        "Avg. Math Score": [avg_math_score],
        "Avg. Reading Score": [avg_read_score],
        "% Students Passing Math": [percent_passing_math],
        "% Students Passing Reading": [percent_passing_read],
        "% Overall Passing": [percent_overall_passing],
    })

district_summary_df

# type(school_data_complete_df["school_name"])
# school_data_complete_df["school_name"].value_counts()
# school_data_complete["student_name"].unique

#sorted(school_data_complete_df.columns)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#call relevant series from relevant dataframe 
#school_summary_df["Per_Student_Budget"]

#sort to see lowest and highest spend
#school_summary_df.sort_values(by="Per_Student_Budget", ascending=True).head(5)
#school_summary_df.sort_values(by="Per_Student_Budget", ascending=False).head(5)

#create new df
scores_by_school_spend_df = pd.DataFrame()

#create 4 relevant bins based on school spend
scores_by_school_spend_df["Spending Range (Per Student)"] = pd.cut(school_summary_df["Per_Student_Budget"], [0, 585, 630, 645, 680], labels=["<$585", "$585-630", "$630-645", "$645-680"])

#add additional relevant series to dataframe
scores_by_school_spend_df["Average Math Score"] = school_summary_df["Average_Math_Score"]
scores_by_school_spend_df["Average Reading Score"] = school_summary_df["Average_Reading_Score"]
scores_by_school_spend_df["% Passing Math"] = school_summary_df["Percent_Passing_Math"]
scores_by_school_spend_df["% Passing Reading"] = school_summary_df["Percent_Passing_Reading"]
scores_by_school_spend_df["Overall Passing Rate"] = school_summary_df["Percent_Overall_Passing"]

#group by spending range
scores_by_school_spend_df.groupby("Spending Range (Per Student)").mean()

```
